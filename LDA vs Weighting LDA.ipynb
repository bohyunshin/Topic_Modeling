## LDA vs Weighting LDA
### Reference
* tomotopy library

```python
import tomotopy as tp
mdl = tp.LDAModel(k=20)
for line in open('sample.txt', encoding='utf-8'):
    mdl.add_doc(line.strip().split())

for i in range(0, 10000, 1000):
    mdl.train(10)
    print('Iteration: {}\tLog-likelihood: {}'.format(i, mdl.ll_per_word))

for k in range(mdl.k):
    print('#############################################################')
    print('Top 10 words of topic {}'.format(k))
    for tup in mdl.get_topic_words(k, top_n=10):
        print(tup)
```

```python
pmi_mdl = tp.LDAModel(k=20, tw = tp.TermWeight.PMI)
for line in open('sample.txt', encoding='utf-8'):
    pmi_mdl.add_doc(line.strip().split())

for i in range(0, 10000, 1000):
    pmi_mdl.train(10)
    print('Iteration: {}\tLog-likelihood: {}'.format(i, pmi_mdl.ll_per_word))

for k in range(pmi_mdl.k):
    print('#############################################################')
    print('Top 10 words of topic {}'.format(k))
    for tup in pmi_mdl.get_topic_words(k, top_n=10):
        print(tup)
```

#### 결과
* weighting을 했음에도 perplexity가 훨씬 나빠졌다??
* 사실 주제별 단어 분포를 보면 불용어 등이 weighting을 했을 때 훨씬 안보였다.
* 추후에 topic coherence까지 비교를 해봐야 겠다.

```python
print('일반 LDA의 perplexity: {}'.format(mdl.perplexity))
print('PMI로 wiehgting한 LDA의 perplexity: {}'.format(pmi_mdl.perplexity))
```
